```python
import pandas as pd
```

### Carregando dadods

```python
data = pd.read_csv('DATASET_interests.csv')
data.sort_values(by = 'text').head(20)
```

### Verificando **duplicatas**

```python
spams = data[data['text'].duplicated(keep = False)]
spams.sort_values(by = 'text').head(15)
```

### Quantos valores duplicados ?

```python
'Total: {}'.format(len(spams))
```

```python
"""
keep :
    - first : Drop duplicates except for the first occurrence.
    - last :  Drop duplicates except for the last occurrence.
    - False : Drop all duplicates.
"""
data.drop_duplicates(subset = 'text', keep = False, inplace = True)
```

```python
data.sort_values(by = 'text').head(15)
```

```python
'Total: {}'.format(len(data))
```

### Matando dado na unha

```python
data[data['user_name'].str.contains("tmj")]
```

```python
data[data['user_name'].str.contains("eq_map")]
```

## Removendo esses cabra

```python
data = data[~ data['user_name'].str.contains("tmj")]
data = data[~ data['user_name'].str.contains("eq_map")]
```

### Quantos exemplos ainda sobraram no conjunto?? 

```python
'Total: {}'.format(len(data))
```

### Exportando os dados

```python
data.to_csv("DATASET_FINAL.csv",  index = None)
```
