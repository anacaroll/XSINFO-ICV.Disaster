```python
import pandas as pd
import folium
from folium.plugins import HeatMap
import pandas as pd
```

### Carregando dados

```python
data_k = pd.read_csv('DATASET_3.csv')
data_k.head()
```

### Visualizando erros

```python
def view_map(data_k):
    posix = data_k.index[0] 
    hmap = folium.Map(location = (data_k['latitude'][posix], data_k['longitude'][posix]), zoom_start = 9)
    hm_wide = hmap.add_child(HeatMap( list(zip(data_k.latitude.values, data_k.longitude.values)),
                       min_opacity=0.5,
                       radius = 11, blur= 15, 
                       max_zoom = 1, 
                     ))
    return hm_wide
```

#### Antes

```python
view_map(data_k)
```

```python
user_names = 'earthshook, Hazards_Network, disasteraware, educaciondecr, dsavior808, MarkNakashima'
```

```python
for index in user_names.split(','):
    data_k = data_k[~ data_k['user_name'].str.contains(index.replace(' ', ''))]
```

#### Depois

```python
view_map(data_k)
```

```python
data_k.to_csv("DATASET_FINAL.csv",  index = None)
```
