```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Carregando dados

```python
data = pd.read_csv('DATASET_1.csv')
data.head()
```

### Carregando **dicion√°rio de palavras**

```python
with open('disasters_words.txt') as loader:
    tmp = loader.readlines()
    text = " ".join(tmp).replace(',',' ').lower().split(' ')
    dict_worlds = sorted(set([x for x in text if len(x) > 1]))
```

### Verificando **Intersec√ß√£o**
![inter](https://upload.wikimedia.org/wikipedia/commons/4/45/Set_intersection.png)

### Adicionando novo campo ao datframe

```python
data['interest'] = 0
data.head(2)
```

```python
for index in data.index:
    text = data['text'][index]
    result = pd.Index(str(text).split()).intersection(pd.Index(dict_worlds)) 
    if len(result) > 0:
        if data['interest'][index] == 1:
            pass
        else:
            for perr in data[data['user_name'].str.contains(data['user_name'][index])].index:
                data.loc[perr, 'interest'] = 1
```

### Verificando o dataframe ap√≥s a execu√ß√£o do algoritmo de sele√ß√£o

```python
relateds = data[data['interest'] == 1]
print("Total de conte√∫do relacionado: {}\n".format(len(relateds)))
data.head(15)
```

### Removendo a coluna **interest**

```python
relateds.drop(columns='interest', inplace = True)
```
