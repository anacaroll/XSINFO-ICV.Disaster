```python
def find_emojis(lista):
    allchars = [str for str in lista.replace(" ","")]
    return [c for c in allchars if c in emoji.UNICODE_EMOJI]

def remove_emoji(string):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

### Frequência de comentários por dias.... as palavras mais frequentes

```python
def frequency_all(qtd_words):
    
    alls_emojis = []
    data_plot = {}
    counter = Counter()
    
    for text in data['text']:
        [alls_emojis.append(x) for x in find_emojis(text)]
        counter.update(text.split())
    
    for tag, count in counter.most_common(qtd_words):
        if len(tag) > 3:
            data_plot.update({tag: count})

    return alls_emojis, data_plot        
            
def plot_words(data_plot, x, y):
    
    qtds = list(data_plot.values())
    words = list(data_plot.keys())
    index = np.arange(len(qtds))

    plt.figure(figsize = (x, y))
    plt.plot(index, qtds, color='r')
    plt.bar(index, qtds)
    plt.xticks(index, words, rotation = 75,fontsize = 15)
    plt.show()
    
def frequency_perday():
    
    data_plot = []
    clusters = data.groupby(by = 'date')
    for volta, date in enumerate(data['date'].unique()):
        counter = Counter()
        flag = [[],[]]

        for text in clusters.get_group(date)['text']:
            words = remove_emoji(text).split()
            counter.update(words)
        for tag, count in counter.most_common():
            if len(tag) > 3:
                flag[0].append(tag)
                flag[1].append(count)
        data_plot.append([date, flag])
    return data_plot

def plot_words_perday(data_plot, qtd_plots = 5, qtd_words = 5):
    count = 0
    fig, axs = plt.subplots(1, qtd_plots, figsize = (15, 4), sharey =True)
    for title, line in data_plot:
        axs[count].bar(line[0][:qtd_words], line[1][:qtd_words])
        axs[count].set_title(title)
        for tick in axs[count].get_xticklabels():
            tick.set_rotation(85)
        count += 1
        if count == qtd_plots:
            break;
```

```python
plot_words_perday(frequency_perday(), 5, 15)
# Os parâmetros correspondem respectivamente a quantidade de plots e  quantidade de keywords
```

### **Wordcount**, verificando quais as keywords com maior importância

```python
emojis, words = frequency_all(60)

plot_words(words, 15, 8)
#São par}ametros de dimensão da imagem
```
