```python
import folium
from folium.plugins import HeatMap
import pandas as pd

data_k = pd.read_csv('DATASET_FINAL.csv')
posix = data_k.index[0] 
hmap = folium.Map(location = (data_k['latitude'][posix], data_k['longitude'][posix]), zoom_start = 9)
hm_wide = hmap.add_child(HeatMap( list(zip(data_k.latitude.values, data_k.longitude.values)),
                   min_opacity=0.5,
                   radius = 11, blur= 15, 
                   max_zoom = 1, 
                 ))
hm_wide
```
